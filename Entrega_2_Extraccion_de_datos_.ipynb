```python
# importar librerias
import pandas as pd
from google.colab import files
import re

# =========================
# 1. Leer archivos desde GitHub
# =========================
url_df1 = "https://raw.githubusercontent.com/Danielavelz/Proyecto---Analitica-de-negocios-I/main/Migracion%20Poblaci%C3%B3n%20total.xls"
url_df2 = "https://raw.githubusercontent.com/Danielavelz/Proyecto---Analitica-de-negocios-I/main/Crecimiento%20del%20PIB%20anual.xls"
```

```python
# Leer archivos
df1 = pd.read_excel(url_df1, skiprows=3) # Se debe cargar primero el archivo de migración
df2 = pd.read_excel(url_df2, skiprows=3) # Se debe cargar luego el archivo de PIB, puesto que el código esta hecho para que sea cargado en ese orden

# Limpiar nombres de columnas con la función
def limpiar_columnas(df):
    nuevas_columnas = {}
    for col in df.columns:
        match = re.search(r'\d{4}', str(col))
        if match:
            nuevas_columnas[col] = match.group() #Deja solo año
        else:
            nuevas_columnas[col] = col
    return df.rename(columns=nuevas_columnas)

# otorgamos nombres a los df limpios
df1 = limpiar_columnas(df1)
df2 = limpiar_columnas(df2)

# Filtrar desde 1990 hasta el último año reportado
def filtrar_desde_1990(df):
    columnas_anios = [col for col in df.columns if str(col).isdigit() and int(col) >= 1990]
    return df[['Country Name'] + columnas_anios]

df1 = filtrar_desde_1990(df1)
df2 = filtrar_desde_1990(df2)

# Renombrar columnas
df1 = df1.rename(columns=lambda x: f"{x}_migracion" if x != 'Country Name' else x)
df2 = df2.rename(columns=lambda x: f"{x}_pib" if x != 'Country Name' else x)

# Unir DataFrames
df_final = pd.merge(df1, df2, on='Country Name', how='inner')

# Resultado
print("Columnas:", df_final.shape[1])
print(df_final.head())

# Exportar base completa
df_final.to_excel("resultado_final.xlsx", index=False)

# Lista de países de Suramérica
paises_suramerica = [
    "Argentina", "Bolivia", "Brazil", "Chile", "Colombia",
    "Ecuador", "Guyana", "Paraguay", "Peru",
    "Suriname", "Uruguay", "Venezuela"
]

# Filtrar DataFrame con los países de Suramérica
df_suramerica = df_final[df_final['Country Name'].isin(paises_suramerica)]

# Exportar a Excel
df_suramerica.to_excel("suramerica_datos.xlsx", index=False)

# Descargar archivo
files.download("suramerica_datos.xlsx")
```
